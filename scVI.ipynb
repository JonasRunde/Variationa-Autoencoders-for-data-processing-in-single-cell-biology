```python
import pandas as pd
import anndata as ad
import numpy as np

#scvi.settings.dl_pin_memory_gpu_training = True

print("imports successful")

adata = ad.read("/content/drive/My Drive/Geneformer/Small.h5ad")

print("Data successfully loaded")

scvi.model.SCVI.setup_anndata(
    adata,
    layer="counts"
)

model = scvi.model.SCVI(adata, n_latent = 256, n_layers = 1, n_hidden = 128)
model

print("model settings generated")

model.train(max_epochs = 600)

print("model train complete")
latent = model.get_latent_representation()
df = pd.DataFrame(latent)

df.to_csv("/content/drive/My Drive/Geneformer/scVI/Small_Latent.csv")

print(df)
```
